```python

import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = sns.load_dataset('iris')
df.head()
```

```python

k_rng = range(1,10)
sse=[]

for k in k_rng:
  km = KMeans(n_clusters=k)
  km.fit(df[[ 'petal_length', 'petal_width']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng, sse)
plt.xlabel('k_rng')
plt.ylabel('Sum of Squared Errors')
plt.title('Sum of Squared Errors vs. k_rng')
plt.grid(True)  # Add gridlines for better readability, optional
plt.show()
```

```python

km = KMeans(n_clusters=3,random_state=0,)
y_predicted = km.fit_predict(df[['petal_length','petal_width']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head(150)
```

```python

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(df.species, df.cluster)
cm
```

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

true_labels = df.species
predicted_labels = df.cluster

cm = confusion_matrix(true_labels, predicted_labels)
class_labels = ['Setosa', 'Versicolor', 'Virginica']

# Plot confusion matrix with black background
plt.figure(facecolor='black')
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Matrix of Confusion', color='white')
plt.colorbar()

tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks, class_labels, color='white')
plt.yticks(tick_marks, class_labels, color='white')

# Fill matrix with values
for i in range(len(class_labels)):
    for j in range(len(class_labels)):
        plt.text(j, i, str(cm[i][j]), ha='center', va='center', color='white')

plt.xlabel('Predicted', color='white')
plt.ylabel('True', color='white')
plt.show()
```
